```python
import heapq
import math

def a_star_graph(graph, coords, start, goal):
    def h(node):
        x1, y1 = coords[node]
        x2, y2 = coords[goal]
        return math.sqrt((x1-x2)**2 + (y1-y2)**2)

    open_set = [(h(start), 0, start, [])]
    visited = {}

    while open_set:
        f, g, current, path = heapq.heappop(open_set)

        if current == goal:
            return path + [current], g

        if current in visited and visited[current] <= g:
            continue
        visited[current] = g

        for neighbor, weight in graph.get(current, {}).items():
            new_g = g + weight
            new_f = new_g + h(neighbor)
            heapq.heappush(open_set, (new_f, new_g, neighbor, path + [current]))

    return None, float('inf')

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}
coords = {'A': (0,0), 'B': (1,0), 'C': (1,1), 'D': (2,1)}

path, cost = a_star_graph(graph, coords, 'A', 'D')
print(f"Duong di: {path} voi tong chi phi: {cost}")
```
