```python
import heapq

class Puzzle4x4:
    def __init__(self, board, g=0, parent=None):
        self.board = board
        self.n = 4
        self.g = g
        self.parent = parent
        self.h = self.calculate_manhattan()
        self.f = self.g + self.h

    def calculate_manhattan(self):
        dist = 0
        for r in range(4):
            for c in range(4):
                val = self.board[r][c]
                if val != 0:
                    target_r, target_c = (val - 1) // 4, (val - 1) % 4
                    dist += abs(r - target_r) + abs(c - target_c)
        return dist

    def __lt__(self, other):
        return self.f < other.f

    def get_neighbors(self):
        neighbors = []
        r, c = next((r, c) for r in range(4) for c in range(4) if self.board[r][c] == 0)
        for dr, dc in [(-1, 0), (1, 0), (0, -1), (0, 1)]:
            nr, nc = r + dr, c + dc
            if 0 <= nr < 4 and 0 <= nc < 4:
                new_board = [list(row) for row in self.board]
                new_board[r][c], new_board[nr][nc] = new_board[nr][nc], new_board[r][c]
                neighbors.append(Puzzle4x4(tuple(map(tuple, new_board)), self.g + 1, self))
        return neighbors

def is_solvable_4x4(board):
    flat = [num for row in board for num in row if num != 0]
    inversions = sum(1 for i in range(len(flat)) for j in range(i + 1, len(flat)) if flat[i] > flat[j])
    empty_row_from_bottom = 4 - next(r for r in range(4) for c in range(4) if board[r][c] == 0)
    return (inversions + empty_row_from_bottom) % 2 != 0 if 4 % 2 == 0 else inversions % 2 == 0

def solve_15_puzzle(start_board):
    start_tuple = tuple(map(tuple, start_board))
    if not is_solvable_4x4(start_board):
        return "Unsolvable"

    goal = tuple(tuple((i*4 + j + 1) % 16 for j in range(4)) for i in range(4))
    open_set = [Puzzle4x4(start_tuple)]
    closed_set = {start_tuple: 0}

    while open_set:
        curr = heapq.heappop(open_set)
        if curr.board == goal:
            path = []
            while curr:
                path.append(curr.board)
                curr = curr.parent
            return path[::-1]

        for neighbor in curr.get_neighbors():
            if neighbor.board not in closed_set or neighbor.g < closed_set[neighbor.board]:
                closed_set[neighbor.board] = neighbor.g
                heapq.heappush(open_set, neighbor)
    return None

start_4x4 = [
    [1, 6, 3, 11],
    [5, 2, 7, 8],
    [9, 10, 4, 12],
    [13, 14, 0, 15]
]
res = solve_15_puzzle(start_4x4)

print(f"Tim thay loi giai sau {len(res)-1} buoc")
for i, step in enumerate(res):
    print(f"Buoc {i}:")
    for row in step:
        print(" ".join(f"{num:2d}" for num in row))
    print("-" * 12)
```
