```python
import heapq

def tsp_a_star(matrix):
    n = len(matrix)
    def h(current, visited):
        remaining = [i for i in range(n) if not (visited & (1 << i))]
        if not remaining: return matrix[current][0]
        return min(matrix[current][i] for i in remaining)
    open_set = [(0, 0, 0, 1, [0])]
    while open_set:
        f, g, u, visited, path = heapq.heappop(open_set)
        if visited == (1 << n) - 1:
            return path + [0], g + matrix[u][0]
        for v in range(n):
            if not (visited & (1 << v)):
                new_g = g + matrix[u][v]
                new_visited = visited | (1 << v)
                heapq.heappush(open_set, (new_g + h(v, new_visited), new_g, v, new_visited, path + [v]))
dist_matrix = [
    [0, 10, 15, 20],
    [10, 0, 35, 25],
    [15, 35, 0, 30],
    [20, 25, 30, 0]
]
path, cost = tsp_a_star(dist_matrix)
print(f"TSP Path: {path}, Cost: {cost}")
```
